```python
def create_plotly (df, metric_name):
    import plotly.graph_objects as go
    
    fig = go.Figure()

    # Add traces
    fig.add_trace(go.Scatter(x=df['bpp_bit_1'], y=df[str(metric_name)+'_1'],
                             mode='markers', 
                             name='bit_1',
                             text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_2'], y=df[str(metric_name)+'_2'],
                            mode='markers',
                            name='bit_2',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_3'], y=df[str(metric_name)+'_3'],
                            mode='markers',
                            name='bit_3',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_4'], y=df[str(metric_name)+'_4'],
                            mode='markers',
                            name='bit_4',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_5'], y=df[str(metric_name)+'_5'],
                            mode='markers',
                            name='bit_5',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_6'], y=df[str(metric_name)+'_6'],
                            mode='markers',
                            name='bit_6',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_7'], y=df[str(metric_name)+'_7'],
                            mode='markers',
                            name='bit_7',
                            text=df['image_no']))

    fig.add_trace(go.Scatter(x=df['bpp_bit_8'], y=df[str(metric_name)+'_8'],
                            mode='markers',
                            name='bit_8',
                            text=df['image_no']))

    fig.update_layout(title='RD curve for '  +  str(metric_name)  + ' quality Metric'
                      ,yaxis_zeroline=False, xaxis_zeroline=False,
                      autosize=False,
                      width=950,
                      height=700,
                     )
    fig.update_xaxes(title='Bit Rate')
    fig.update_yaxes(title = str(metric_name))
    fig.show()
```

```python
def plotly_graph(bpp_dict, metric_dict, metric_name):
    import pandas as pd
    '''
    @bpp_dict = Dictionary of bit rate of the image
    @metric_dict = Dictionary of quality metric of the image
    @metric_name = Metric name in string format
    
    '''
    metric= pd.DataFrame(metric_dict).T
    metric['image_number']= ['img'+str(i) for i in range (1,601)]
    metric = metric.set_axis([str(metric_name)+'_1', str(metric_name)+'_2', str(metric_name)+'_3', str(metric_name)+'_4', str(metric_name)+'_5',str(metric_name)+'_6',str(metric_name)+'_7',str(metric_name)+'_8','image_no'], axis=1, inplace=False)
    
    bpp= pd.DataFrame(bpp_dict).T
    bpp['image_number']= ['img'+str(i) for i in range (1,601)]
    bpp = bpp.set_axis(['bpp_bit_1', 'bpp_bit_2', 'bpp_bit_3', 'bpp_bit_4', 'bpp_bit_5','bpp_bit_6','bpp_bit_7','bpp_bit_8','image_no'], axis=1, inplace=False)
    
    df = pd.merge(bpp, metric)
    
    g= create_plotly(df, metric_name)
    
    return g
    
```

```python
plotly_graph(dbpp_, dpsnr_, 'PSNR')
plotly_graph(dbpp_, dmsssim_, 'MSSSIM')
plotly_graph(dbpp_, dfsim_, 'FSIM')
plotly_graph(dbpp_, dvif_, 'VIF')
```
